```kotlin
import com.github.ax_as.datapro.data.Convocador
import com.github.ax_as.datapro.data.DataFiles
import com.github.ax_as.datapro.data.GerenciadorConvocacoes
import com.github.ax_as.datapro.data.ListaAprovados
import com.github.ax_as.datapro.data.add_prova_pratica_final
import com.github.ax_as.datapro.data.add_prova_pratica_prov
import com.github.ax_as.datapro.data.avaliacaoSerpro
import com.github.ax_as.datapro.data.concursoSerproObjetivas
import com.github.ax_as.datapro.data.final_objetivas
import com.github.ax_as.datapro.data.final_objetivas_pond
import com.github.ax_as.datapro.data.headerPraticaProv
import com.github.ax_as.datapro.data.manterClassificados
import com.github.ax_as.datapro.data.praticaSerpro
import com.github.ax_as.datapro.data.removerZeradosPratica
import com.github.ax_as.datapro.data.toDefaultHtml
import java.nio.file.Path

fun toLocalPath(file: String) = Path.of("/data/files/output/$file")

val configuration = DisplayConfiguration(
    rowsLimit = null,
    useDarkColorScheme = true,
)

var listaAC =
    DataFrame.concursoSerproObjetivas(DataFiles.objetivas_ac)
        .final_objetivas("AC")
        .final_objetivas_pond()


var listaPPP =
    DataFrame.concursoSerproObjetivas(DataFiles.objetivas_ppp)

        .final_objetivas("PPP")
        .final_objetivas_pond()

var listaPCD =
    DataFrame.concursoSerproObjetivas(DataFiles.objetivas_pcd)
        .final_objetivas("PCD")
        .final_objetivas_pond()


var pratica_ac =
    DataFrame.praticaSerpro(DataFiles.serpro_final_pratica, "AC")
        .add("pos_pratica") { index() + 1 }


var pratica_ppp =
    DataFrame.praticaSerpro(DataFiles.serpro_final_pratica_ppp, "PPP")
        .add("pos_pratica") { index() + 1 }

var pratica_pcd =
    DataFrame.praticaSerpro(DataFiles.serpro_final_pratica_pcd, "PCD")
        .add("pos_pratica") { index() + 1 }

var provisoria_ac =
    DataFrame.praticaSerpro(
        DataFiles.serpro_prov_pratica,
        "AC",
        header = headerPraticaProv
    )
        .add("pos_pratica") { index() + 1 }

var provisoria_pcd =
    DataFrame.praticaSerpro(
        DataFiles.serpro_prov_pratica_pcd,
        "PCD",
        header = headerPraticaProv
    )
        .add("pos_pratica") { index() + 1 }
var provisoria_ppp =
    DataFrame.praticaSerpro(
        DataFiles.serpro_prov_pratica_ppp,
        "PPP",
        header = headerPraticaProv
    )
        .add("pos_pratica") { index() + 1 }


var acComPratica = listaAC
    .add_prova_pratica_prov(provisoria_ac)
    .add_prova_pratica_final(pratica_ac)
    .select(
        "pos_final", "lista", "inscricao", "nome",
        "p1",
        "accb",
        "p2",
        "acce",
        "NF_OBJ",
        "p3",
        "NFC"
    ).removerZeradosPratica()

var avaliacaoSerproPcd = DataFrame.avaliacaoSerpro(DataFiles.serpro_avaliacao_pcd)
var avaliacaoSerproPPP = DataFrame.avaliacaoSerpro(DataFiles.serpro_avaliacao_ppp)

avaliacaoSerproPcd.toDefaultHtml().openInBrowser()
var pppComPratica = listaPPP
    .add_prova_pratica_prov(provisoria_ppp)
    .add_prova_pratica_final(pratica_ppp)
    .manterClassificados(avaliacaoSerproPPP)
    .select(
        "pos_final", "lista", "inscricao", "nome",
        "p1",
        "accb",
        "p2",
        "acce",
        "NF_OBJ",
        "p3",
        "NFC"
    ).removerZeradosPratica()

var pcdComPratica = listaPCD
    .add_prova_pratica_prov(provisoria_pcd)
    .add_prova_pratica_final(pratica_pcd)
    .manterClassificados(avaliacaoSerproPcd)
    .select(
        "pos_final", "lista", "inscricao", "nome",
        "p1",
        "accb",
        "p2",
        "acce",
        "NF_OBJ",
        "p3",
        "NFC"
    ).removerZeradosPratica()

var convocacoes = DataFrame.empty()
val primeiraPpp = 3
val primeiraPcd = 5
val primeiraAc = 1

val listaPpp = ListaAprovados(primeiraPpp, 5, pppComPratica)
val listaPcd = ListaAprovados(primeiraPcd, 5, pcdComPratica)
val listaAc = ListaAprovados(candidatos = acComPratica)


val excluidosPPP = listaPPP.join(avaliacaoSerproPPP, JoinType.Exclude){
    it["nome"] match right ["nome"]
}
excluidosPPP.toDefaultHtml().openInBrowser()

val excluidosPCD = listaPCD.join(avaliacaoSerproPcd, JoinType.Exclude){
    it["nome"] match right ["nome"]
}
excluidosPCD.toDefaultHtml().openInBrowser()

val g = GerenciadorConvocacoes<Convocador<*>>(
    arrayListOf(
        listaPpp,
        listaPcd,
        listaAc
    )
)

var vaga = 1
val conv = mutableListOf<DataRow<*>>()
while (true) {
    val candidato = g.convocar(vaga)
    if (candidato == Convocador.emptyResponse) {
        break
    }
    val rowCandidato = candidato as DataRow<*>
    conv.add(rowCandidato)

    vaga++
}

val df = conv.toDataFrame().distinctBy("nome")
    .add("vaga") { index() + 1 }.moveToLeft("vaga")

//    .select("lista", "inscricao", "nome", )

df.toDefaultHtml().writeHTML(toLocalPath("ordem_convocacoes_apos_avaliacoes.html"))




```
